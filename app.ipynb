```python
from flask_restful import Resource,Api
from flask import Flask , render_template , request
from flask_sqlalchemy import SQLAlchemy
import os

app = Flask(__name__)
app.config['SQLALCHEMY_DATABASE_URI'] = "sqlite:///bot_sqlite.sqlite3"

db = SQLAlchemy(app)
api = Api(app)

class Bank(db.Model):
    ifsc = db.Column(db.String,primary_key=True)
    bank_id = db.Column(db.Integer)
    branch = db.Column(db.String)
    address = db.Column(db.String)
    city = db.Column(db.String)
    district = db.Column(db.String)
    state = db.Column(db.String)
    bank_name = db.Column(db.String)

import sqlite3
class BANK_SEARCH(Resource):
    def get(self):
        connection = sqlite3.connect(r"C:\Users\athik\OneDrive\Desktop\Bot\instance\bot_sqlite.sqlite3")
        posts = connection.execute('SELECT * FROM bank').fetchall()
        connection.close()
        query = request.args.get('q').upper()
        limit = int(request.args.get('limit'))
        offset = int(request.args.get('offset'))
        res = []
        
        for i in posts:
            flag=0
            for j in range(2,8):
                if i[j] and query in i[j]:
                    flag=1
            if flag:
                res.append(i)

    
        obj = res[(offset+1):(offset+limit+1)]
        b=[]
        for i in obj:
            b.append((i[0],{"ifsc":i[0],
                              "bank_id":i[1],
                              "branch":i[2],
                              "address":i[3],
                              "city":i[4],
                              "district":i[5],
                              "state":i[6],
                              "bank_name":i[7]
                              }))
       
            
        b.sort()
        temp=[]
        for i in b:
           temp.append(i[1])
        return {
            "branches":temp
        }

class BANK_BRANCH(Resource):
    def get(self):
        connection = sqlite3.connect(r"C:\Users\athik\OneDrive\Desktop\Bot\instance\bot_sqlite.sqlite3")
        posts = connection.execute('SELECT * FROM bank').fetchall()
        connection.close()
        query = request.args.get('q').upper()
        limit = int(request.args.get('limit'))
        offset = int(request.args.get('offset'))
        res = []
        
        for i in posts:
            flag=0
            for j in range(2,8):
                if i[j] and query in i[j]:
                    flag=1
            if flag:
                res.append(i)

    
        obj = res[(offset+1):(offset+limit+1)]
        b=[]
        for i in obj:
            b.append((i[0],{"ifsc":i[0],
                              "bank_id":i[1],
                              "branch":i[2],
                              "address":i[3],
                              "city":i[4],
                              "district":i[5],
                              "state":i[6],
                              "bank_name":i[7]
                              }))
       
            
        b.sort(reverse=True)
        temp=[]
        for i in b:
           temp.append(i[1])
        return {
            "branches":temp
        }

api.add_resource(BANK_SEARCH,'/api/search')
api.add_resource(BANK_BRANCH,'/api/branch')

if __name__ == '__main__':
    app.run(port=5050)
```
